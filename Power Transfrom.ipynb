```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import scipy.stats as stats
```

```python
df = pd.read_csv('GoogleStock Price.csv',usecols = ['Open','High','Low','Close','Adj Close','Volume'])

x= df[['Open','High','Low','Close','Adj Close']]
y= df['Volume']

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)


for col in X_train:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(X_train[col], dist="norm", plot=plt)
    plt.title(col)

    plt.show()
```

```python
#using Box-cox transform
pt = PowerTransformer(method ='box-cox')
pt.fit(X_train)
X_train_box = pt.fit_transform(X_train)
X_train_box = pd.DataFrame(X_train_box)

for col in X_train_box.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train_box[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(X_train_box[col], dist="norm", plot=plt)
    plt.title(col)

    plt.show()
```

```python
#Using Yeo-Johnson transform
pt = PowerTransformer()
pt.fit(X_train)
X_train_yon = pt.fit_transform(X_train)
X_train_yon = pd.DataFrame(X_train_box)

for col in X_train_box.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train_yon[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(X_train_yon[col], dist="norm", plot=plt)
    plt.title(col)

    plt.show()
```
