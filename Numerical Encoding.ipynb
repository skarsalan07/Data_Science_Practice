```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
import matplotlib.pyplot as plt 
```

```python
#Using Uniform 
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
df.dropna(inplace = True)
x = df.iloc[:,1:]
y = df["Survived"]

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
Kbin_age = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='quantile')
Kbin_fare = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='quantile')

trf = ColumnTransformer([
    ('first',Kbin_age,[0]),
    ('second',Kbin_fare,[1])
])

X_train_trf = trf.fit_transform(X_train)
output = pd.DataFrame({
    'age':X_train['Age'],
    'age_trf':X_train_trf[:,0],
    'fare':X_train['Fare'],
    'fare_trf':X_train_trf[:,1]
})
output
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(output['age'])
plt.title("Before")
plt.subplot(122)
plt.hist(output['age_trf'],color='red')
plt.title("After")
plt.show()

plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(output['fare'])
plt.title("Before")
plt.subplot(122)
plt.hist(output['fare_trf'],color='red')
plt.title("Fare")
plt.show()
 
```
