```python
#Standardization mathod 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
df= pd.read_csv('social network ads.csv')

#it's (optional)... split data into train and test split and then process for standardization
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size = 0.3,random_state=0)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns )
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns )

np.round(X_train_scaled.describe(),1)
```

```python
#Effect after Standardization technique is applied
plt.title("Before standardization is applied")
plt.scatter(X_train['Age'],X_train['EstimatedSalary'])
plt.show()
plt.title("before standardization is applied")
plt.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],color = "red")
plt.show()
```
