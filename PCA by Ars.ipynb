```python
import plotly.express as px
import pandas as pd
import numpy  as np
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("social network ads.csv")
ss = StandardScaler()
standard_df = ss.fit_transform(df[['Age','EstimatedSalary','Purchased']])
standard_df = pd.DataFrame(standard_df,columns = ['Age','EstimatedSalary','Purchased'])
standard_df
```

```python
cov = np.cov(standard_df,rowvar=False)

eig_val,eig_vec = np.linalg.eig(cov)

pc= eig_vec[:, :2]
pc.sort()
transformed_df = np.dot(standard_df,pc)
df = pd.DataFrame(transformed_df,columns = ['Pc1','Pc2'])
eig_val
print(eig_vec,eig_val)
```
