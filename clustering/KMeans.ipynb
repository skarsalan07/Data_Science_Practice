```python
 import pandas as pd
 import numpy as np
```

```python
data = pd.read_csv('/content/Data_Science_Practice/clustering/student_clustering.csv')
data
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(data)
  wcss.append(kmeans.inertia_)
```

```python
from matplotlib import pyplot as plt
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
k = KMeans(n_clusters=4,init='k-means++',random_state=42)
y_kmeans = k.fit_predict(data)
```

```python
plt.scatter(data['iq'],data['cgpa'])
```

```python
# prompt: using above code plat that clusters into different colours

plt.scatter(data.iloc[y_kmeans == 0, 0], data.iloc[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(data.iloc[y_kmeans == 1, 0], data.iloc[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(data.iloc[y_kmeans == 2, 0], data.iloc[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(data.iloc[y_kmeans == 3, 0], data.iloc[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(k.cluster_centers_[:, 0], k.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('IQ')
plt.ylabel('CGPA')
plt.legend()
plt.show()
```
